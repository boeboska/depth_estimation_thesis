```python
def remove_first_col(data):
    x_axis = []
    y_axis = []
    for x in data:

        x_axis.append(x[2])
        y_axis.append(x[1])

    return x_axis, y_axis
```

```python
def create_empty_dict():
    data_dict = {}

    experiments = ['experiment_13_', 'experiment_14_', 'experiment_20_', 'experiment_18_', 'experiment_21_', 'experiment_26_', 'experiment_22_', 'experiment_27_', 'experiment_28_']
    epochs = ['0', '1', '2', '3', '4', '5']
    loss_types = ['loss', 'loss_within_attention_mask', 'loss_within_attention_mask_dialation_1', 'loss_within_attention_mask_dialation_3']

    for exp in experiments:
        data_dict[exp] = {}
        for epoch in epochs:
            data_dict[exp][epoch] = {}
            for loss_type in loss_types:
                data_dict[exp][epoch][loss_type] = {}

    return data_dict
```

```python
empty_dict = create_empty_dict()
experiments = ['experiment_13_', 'experiment_14_', 'experiment_20_', 'experiment_18_', 'experiment_21_', 'experiment_26_', 'experiment_22_', 'experiment_27_', 'experiment_28_']
epochs = ['0', '1', '2', '3', '4', '5']
loss_types = ['loss', 'loss_within_attention_mask', 'loss_within_attention_mask_dialation_1', 'loss_within_attention_mask_dialation_3']

numers_exp_20 = []
numbers_exp_21 = []
numbers_exp_22 = []


for exp in experiments:

    print("exp", exp)
    for epoch in epochs:
        print("epoch", epoch)
        for loss_type in loss_types:
            print("loss type", loss_type)

            current_data = []

            file_name = exp + 'weights_' + epoch + '_batch_idx9999' + '.pkl'
            with open('../monodepth2/validation_all/' + file_name, 'rb') as f:
                current_exp = pickle.load(f)
                for key, value in current_exp.items():

                    for number in value[loss_type]:
                        if number > 0:
                            current_data.append(number)

                end_number = np.average(np.array(current_data))

                empty_dict[exp][epoch][loss_type] = end_number

for x , y in empty_dict.items():
    print(x,y)
print(empty_dict)
                
```

```python
loss_types = ['loss', 'loss_within_attention_mask', 'loss_within_attention_mask_dialation_1', 'loss_within_attention_mask_dialation_3'] 



epoch = [1, 2, 3, 4, 5]
experiments = [ ['experiment_13_', 'experiment_14_', 'experiment_20_'] , ['experiment_18_', 'experiment_21_', 'experiment_26_'], ['experiment_22_', 'experiment_27_', 'experiment_28_']]

fig, axs = plt.subplots(2, 2, figsize = (18, 10), constrained_layout=True)
# fig.tight_layout()

# subplots_adjust(left=None, bottom=5, right=None, top=None, wspace=None, hspace=None)

for loss_type in loss_types:
    print(loss_type)
    plot_data = []
    curve_data = []
    
    for exp in experiments:
        for run in exp:

            curr = []
            for key, value in empty_dict[run].items():
#                 print(key, value)
 
                curr.append(value[loss_type])
            curve_data.append(curr)


        temp = []
        for x, y, z in zip(curve_data[0], curve_data[1], curve_data[2]):
            temp.append((x + y + z) / 3 )
        plot_data.append(temp)

        curve_data = []

    regular_train_loss_y = [1, 2, 3, 4, 5, 6]

    plt.figure(figsize=(10,5))  
    
    if loss_type == 'loss':
        loss_type = 'Total loss'
        l, r = 0, 0
    if loss_type == 'loss_within_attention_mask':
        loss_type = 'Loss within attention mask'
        l, r = 0, 1
    if loss_type == 'loss_within_attention_mask_dialation_1':
        loss_type = 'Loss within attention mask dialation 1'
        l, r = 1, 0
    if loss_type == 'loss_within_attention_mask_dialation_3':
        loss_type = 'Loss within attention mask dialation 3'
        l, r =1,1

    
    axs[l, r].plot(regular_train_loss_y, plot_data[0], color = 'lightblue', marker = 'd',linewidth=3, label = 'Baseline')
    axs[l, r].plot(regular_train_loss_y, plot_data[1], color = '#ffa64d', marker = 'd',linewidth=3, label = 'Experiment 1')
    axs[l, r].plot(regular_train_loss_y, plot_data[2], color = '#ff80ff', marker = 'd',linewidth=3, label = 'Experiment 2')
    axs[l, r].set_title(f'{loss_type}', fontsize = 20)
    axs[l, r].legend(loc = "upper right", fontsize=10)

for ax in axs.flat:
    ax.set(xlabel='Epoch', ylabel='Loss')

    ax.xaxis.get_label().set_fontsize(16)
    ax.yaxis.get_label().set_fontsize(16)
fig.subplots_adjust(left=None, bottom=None, right=None, top=None, wspace=None, hspace=.4)

plt.show()   
    
```
